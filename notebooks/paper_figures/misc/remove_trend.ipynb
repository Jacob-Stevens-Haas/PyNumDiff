```python
x = np.linspace(0, 100, 1000)
```

```python
y = np.sin(x) + x
```

```python
plt.plot(x,y)
```

```python
import scipy.stats
```

```python
r = scipy.stats.linregress(x, y)
```

```python
r
```

```python
y_trend = x*r.slope + r.intercept
```

```python
plt.plot(x, y_trend)
```

```python
plt.plot(x, y - y_trend)
```
